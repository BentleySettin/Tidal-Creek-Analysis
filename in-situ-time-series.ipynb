# Plot Smith Creek Sensor Data
- Read in In-Situ sensor data
- Read in ILM airport weater data

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.dates import DateFormatter

import pandas as pd
```

### Read in In-Situ data
- First had to find and replace double quotes (") with nothing in VSCode
- Remove note at very end/bottom of file after data

```python
filename = 'VuSitu_2024-02-21_12-51-36_Smith Creek_Log_Smith Creek 23rd St.csv'
filepath = os.path.join('Data', 'InSitu', filename)

df_full = pd.read_csv(filepath, header = 22, parse_dates = [0])
df_full.head()
```

```python
df_full['Datetime'] = pd.to_datetime(df_full['Date Time'])
df_full = df_full.set_index('Datetime')
df_full.head()
```

### Read in weather data
- Download: Station Name, Geographic Location, Include Data Flags, use metric option
- Don't remove quotation marks in VS code. Pandas can handle.

```python
filename = '3653978_ILM_weather.csv'
filepath = os.path.join('Data', 'InSitu', filename)

df_met = pd.read_csv(filepath, header = 0, parse_dates = [5])
df_met.head()
```

### Plot

```python
fig, axs = plt.subplots(3, 1, figsize = (10, 10), sharex = True)
axs[0].plot(df_full.index, df_full.loc[:, 'Pressure (psi) (1099450)'], 'k')
axs[0].set_ylabel('Pressure (psi)')

axs[1].plot(df_full.index, df_full.loc[:, 'Temperature (°C) (1099450)'], 'k')
axs[1].set_ylabel('Temperature (°C)')

axs[2].plot(df_full.index, df_full.loc[:, 'Salinity (PSU) (1099450)'], 'k')
axs[2].set_ylabel('Salinity (PSU)')

# Add precip in a second y-axis
ax2 = axs[2].twinx()
ax2.plot(df_met['DATE'], df_met['PRCP'], 'b--')
ax2.set_ylabel('Daily Precip (mm)', color='b')
ax2.tick_params('y', colors='b')

axs[0].xaxis_date() # make sure it knows that x is a date/time
axs[0].set_xlim(df_full.index.min(), df_full.index.max())

for axi in axs.flat:
#     axi.xaxis.set_major_locator(plt.MaxNLocator(3))
#     print(axi)
    axi.yaxis.set_major_locator(plt.MaxNLocator(5))
#     axi.yaxis.set_major_formatter(ticker.FormatStrFormatter("%.02f"))

# xtick dates are poorly spaced
axs[0].xaxis.set_major_locator(ticker.MultipleLocator(3.5))

date_form = DateFormatter("%m/%d")
axs[0].xaxis.set_major_formatter(date_form)


fig.autofmt_xdate() # makes the date labels easier to read.
plt.tight_layout()
fig.align_ylabels()

# plt.savefig('n23rd.png', dpi = 300)
```
